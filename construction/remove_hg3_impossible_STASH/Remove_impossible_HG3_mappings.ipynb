```python

with open('stash_usage.txt', 'w') as fh:
    for s in sorted(stash_used):
        for k in sorted(stash_used[s]):
            fh.write("\t".join([s, k, ",".join(sorted(list(stash_used[s][k])))]) + "\n")
```
