```python
from dataclasses import dataclass, field
from collections import defaultdict
from textwrap import dedent

TEMPLATE = '''
## Data Request information

| Field | Value | Notes |
| --- | --- | --- |
{dr}


## Mapping information 

| Field | Value | Notes |
| --- | --- | --- |
{mapping}


## STASH entries (relevant for UM only)

| Model | STASH | Section, item number | Time Profile | Domain Profile | Usage Profile |
| --- | --- | --- | --- | --- | --- |
{stash}

'''

@dataclass
class DRIssue:
    """
    Data class to hold issue information
    """
    dr_info: dict = field(default_factory=dict)
    dr_notes: dict = field(default_factory=dict)
    mapping_info: dict = field(default_factory=dict)
    mapping_notes: dict = field(default_factory=dict)
    stash: defaultdict(dict) = field(default_factory=lambda: defaultdict(list))

    def write(self):
        """
        return the issue as text
        """
        dr = '\n'.join(['| {} | {} | {} |'.format(k, self.dr_info[k], self.dr_notes[k]) for k in self.dr_info])
        mapping = '\n'.join(['| {} | {} | {} |'.format(k, self.mapping_info[k], self.mapping_notes[k]) for k in self.mapping_info])
        stash = ''
        for key, stashdata in self.stash.items():
            for entry in stashdata:
                stash += '| {} | {} |\n'.format(key, ' | '.join(entry))

        return TEMPLATE.format(dr=dr, mapping=mapping, stash=stash).replace("  ", " ")

    def write_file(self, filename):
        """
        write issue to file
        """
        with open(filename, "w") as fh:
            fh.write(self.write())

    def read(self, filename):
        """
        read an issue into the class from a file
        """
        with open(filename) as fh:
            lines = [i.strip() for i in fh.readlines()]
        
        skip_lines = ['| Field', '| Model |', '| ---']
        section = None

        for line in lines:
            if not line:
                #skip blank
                continue
            elif any([line.startswith(i) for i in skip_lines]):
                # skip table headers
                continue
            elif line.startswith('## Data Request information'):
                section = 'DR'
                continue
            elif line.startswith('## Mapping information'):
                section = 'M'
                continue
            elif line.startswith('## STASH entries'):
                section = 'S'
                continue
            
            linedata = [i.strip() for i in line.strip('|').split('|')]
            if section in ['DR', 'M']:
                try:
                    key, value, note = linedata
                except ValueError:
                    print(linedata)
                    print(line)
                    raise
                if section == 'DR':
                    self.dr_info[key] = value
                    self.dr_notes[key] = note
                elif section == 'M':
                    self.mapping_info[key] = value
                    self.mapping_notes[key] = note
            elif section == 'S':
                model = linedata[0]
                stashdata = linedata[1:]
                self.stash[model].append(stashdata)
        return self
            
    
```

```python
HEADING_MAPPING = {
    'Branded variable name': 'branded_variable_name',
    'CF standard name': 'standard_name',
    'Long name': 'long_name',
    'Comment': 'comment',
    'Dimensions': 'dimensions',
    'Frequency': 'frequency',
    'Units': 'units',
    'Positive': 'positive',
    'Processing notes': 'processing_note',
    'Cell methods': 'cell_methods',
    'Cell measures': 'cell_measures',
    'Modeling realm': 'modeling_realm',
    'Variable status': 'variable_status',
}

def update_issue(filename):
    variable = filename.split("/")[-1]
    i = DRIssue().read(filename)
    j = dr122[variable]

    for ki, kj in HEADING_MAPPING.items():
        if i.dr_info[ki] != j[kj]:
            i.dr_notes[ki] += "Updated v1.2.2"
        i.dr_info[ki] = j[kj]
    
    i.write_file(filename)
    
```

```python
done = []
```

```python
import os
issues = os.listdir("../issues")


for issue in issues:
    filename = '../issues/'+issue
    if filename not in done:
        print(filename)
        update_issue(filename)
        done.append(filename)
```
